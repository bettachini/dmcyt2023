## Tendencia al clustering
- Sin etiquetas $\rightarrow$ no supervisado / validación interna
- Con etiquetas $\rightarrow$ supervisado / validación externa

### Coeficiente de Hopkins
$$
H = \frac{\sum_i^p w_i}{\sum_i^p w_i + \sum_i^p u_i} ,
$$
siendo $w_i$ las distancias al vecino más cercano, e __inventando__ puntos que no son del conjunto de datos, se calcula las distancias a estos puntos $u_i$.
Se asume que estos agregados se distribuirán en forma homogénea.

## Validación externa
###  Armado de la matríz de confusión
- No tiene porque ser cuadrada, es decir que no coincida $k$ de clustering que el número de etiquetas (labels).
- En el ejemplo _Cluster_ se refiere a un número secuencial identificatorio. En este ejemplo hay 5 clusters numerados del 0 al 4.
- Se cuentan cuantos casos de una determinada etiqueta se incluyeron en determinado cluster.

### Medida de van Dongen
- Esencialmente cuenta cuantos se van de la "diagonal". No es que se diagonalize ni se orden, esto se refiere a que sopesa cuantos no están en el grupo mayoritario.
- La $\sum_i$ se realiza sobre las $i$ filas, y $\sum_j$ sobre las $j$ columnas.
- Mejor cuanto más próximo sea a $0$

$$
VD_n = \frac{2 n - \sum_i \text{máx}_i n_{ij} - \sum_j \text{máx}_j n_{ij}}{2 n - \text{máx}_i n_i - \text{máx}_j n_j}
$$

### Índice de Rand
Es una alternativa a tener que generar la matríz de confusión antes vista.

- Genero todos los pares posibles de los $n$ elementos
- Se los clasifica según si
  - son de la misma clase
  - están en un mismo cluster
  - esto resulta en cuatro posibilidades: a, b , c, d

| | mismo cluster | distinto cluster |
|----|:------------:|:------------------:|
| misma clase | a  |         c         |
| distinta clase | d  |         b         |

$$
R
= \frac{a+b}{a+b+c+d}
= \frac{a+b}{{n}\choose{2}}
$$

## Validación interna
Coeficientes que sopesan la cohesión y la separación

### Coeficiente de Silhoutte
$$
s_i = \frac{b_i - a_i}{\mathrm{máx}(a_i, b_i)}
$$

- $a_i$ (cohesión) distancia promedio a elementos de su propio cluster.
- $b_i$ (separación) distancias a los centroides de los otros clusters

Valores
- El mejor silhoutte posible es 1 ($a_i = 0)$
- En tanto que $-1$ es cuando no hay (casi) distancia a los otros clusters

### Clustering jerárquico
- Compara las distancias dentro del árbol con las distancias originales.
- Si son similares, el árbol es bueno.
- Tras generar ambas matrices de distancias se calcula la correlación entre ambas.
